# Getting Event Information

This mini-project is all about getting information from the TicketMaster API and converting it into a useable dataframe for future data exploration or processing projects.

I make a call to the TicketMaster API and iterate over the data provided to populate a dataframe with the genre I'm interested in finding information on.

```python
import ticketpy
import pandas as pd
tm_client = ticketpy.ApiClient('Key_Name_Here')

pages = tm_client.events.find(
    classification_name='Electronic',
    state_code='CA'
)
eventlist = []
for page in pages:
    for e in page:
        event = e.json
    ###getName
        name = event['name']
    ###get Time Start 
        start_time =event['sales']['public']['startDateTime']
    ###get Time End
        end_time = event['sales']['public']['endDateTime']
    ###get event type i.e. music
    ###should fix to have subclasses and other classes attached
        event_type = event['classifications'][0]['segment']['name']
    ###get Address
        address = event['_embedded']['venues'][0]['location']
    ###get Event Description
        if 'info' in event.keys():
            description = event['info']
        else:
            description = 'None'
        listx = [name,start_time,end_time,event_type,address,description]
        eventlist.append(listx)
        
df=pd.DataFrame(eventlist,columns = ['Name','Start Time','End Time','Event Type', 'Address','Description'])
        
        
df
```
