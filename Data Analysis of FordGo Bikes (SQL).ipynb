# The Importance of Weekend Trips

## Executive Summary

##### Overview, Recommendations and Impact
In this report, I propose a solution to the issue of decreased ridership during the weekends. In this report you will find a dive into the breakdown of ridership by day and the differences in ridership on weekends by subscriber type: customer (24hour/3day pass) or subscriber (30 day/annual pass). 

Based on these differences, this report proposes the increasing marketing materials on fun weekend get aways, but also substantially proposed creative discounts for future subscription purchases to subscribers who take trips on weekends, delivered via the mobile app. Secondly, it proposes introducing a weekend flex pass that would remove the 30minute ride time cap on weekends and would offer unlimited rides on weekends.

Evaluating the potential impact of these proposals, if the company managed to increase ridership levels on weekends to be on par with weekdays, this would correspond over our service, to an additional 200,000 trips.

## Problem Statement
This report finds that there is a significant drop in ridership on weekends as shown by the graph generated below.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
! bq query --use_legacy_sql=false --format=csv 'SELECT COUNT(*) as num_trips, (EXTRACT(dayofweek FROM start_date)) as day FROM `bigquery-public-data.san_francisco.bikeshare_trips` GROUP BY (EXTRACT(dayofweek FROM start_date))' > daycounts.csv
df_week = pd.read_csv('daycounts.csv')
df_week = df_week.sort_values(by = ['day'])
df_week = df_week.reset_index(drop = True)
arr_week = np.array(df_week['num_trips'])


#makes graph of rides by weekday
N = len(arr_week)
bottom = 50000

theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = arr_week
width = (2*np.pi) / N

ax = plt.subplot(111, polar=True)
bars = ax.bar(theta, radii, width=width, bottom=bottom)

# Use custom colors and opacity
for r, bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.jet(r / 300000))
    bar.set_alpha(0.8)
radians = np.linspace(0, 2 * np.pi, num=7, endpoint=False)
ax.set_theta_zero_location('E')
    # Set up labels
ax.set_xticks(radians)
ax.set_xticklabels(['Sa','F', 'Th', 'W', 'Tu', 'M', 'Su'], size=14, weight='bold')
ax.set_yticklabels(['0','50', '100', '150'])
ax.set_xlabel('Total Number of Bike Trips by Day of Week (In Thousands)')

plt.show()
```

If you notice here, the number of trips during the weekend is significantly lower than the number of trips on weekdays. This is a big gap in the number of trips taken on these two days. I propose putting incentives in place to increase the amount of weekend trips.

Upon further exploration of this phenomenon, I discovered that the breakdown of subscriber type (subscriber: annual or 30day, customer (24 hour or 3 day pass) for trips differs between week days and weekends:

```python
! bq query --use_legacy_sql=false --format=csv 'SELECT COUNT(*) as num_trips, (EXTRACT(dayofweek FROM start_date) BETWEEN 2 AND 6) as Weekday, subscriber_type FROM `bigquery-public-data.san_francisco.bikeshare_trips` GROUP BY (EXTRACT(dayofweek FROM start_date) BETWEEN 2 AND 6), subscriber_type' > daycounts_sub.csv
df_week_sub = pd.read_csv('daycounts_sub.csv')
df_week_sub = df_week_sub.sort_values(by = ['Weekday', 'subscriber_type'])
df_week_sub = df_week_sub.reset_index(drop = True)
display(df_week_sub)
```

This lookup table succinctly communicates the difference in number of trips by weekday and subscriber type.

From this chart here, you can see a pattern emerge for what kind of people take trips depending on the day of the week. On Weekdays, it's primarily subscribers, but on weekends, it's about 50-50. 

*Simply put, there is a huge pool of subscribers who seemingly are not as actively using their subscription on weekends as much as they are on weekdays. There is a smaller group of customers on weekdays who could also contribute to trips on weekends.*

It is for this reason I would recommend that we can increase ridership by targetting commuters and finding ways to get them to ride more on weekends. To do so, we have access to providing discounts, marketing and promotions via the mobile app users use to boost weekend traffic. However, subscribers have already paid for a year in advance or paid for the month. To address the need to mobilize subscribers on weekends requires creative incentive making.

### My proposal is two fold: 
1) Promote subscriber weekend travel through mobile ads about destinations in their home city explorable on weekends. Provide discounts on the app for monthly subscribers (30 day prepaid) for credits toward next month's subscription fee for taking weekend rides. These discounts would be delivered via the mobile app notification system.

2) Introduce a weekend flex pass, marketed heavily toward both weekday and weekend customers. This weekend pass would offer unlimited rides for the weekend only. There is room for debate and further analysis on consumer propensity to purchase passes on what to price the pass, but for the purpose of increasing ridership, this report would advocate for it to be low, but economically sustainable for the company.

Proposal Part 1 would be highly effective. Specifically, targetting commuters during the weekday could result in massive gains to ridership on weekends!

```python
#estimate the number of commuter trips per weekday
! bq query --use_legacy_sql=false --format=csv 'SELECT subscriber_type, COUNT(*)/5 as number_of_commuter_trips_per_weekday FROM `bigquery-public-data.san_francisco.bikeshare_trips` WHERE (start_station_id != end_station_id) AND ((EXTRACT(HOUR FROM start_date) BETWEEN 4 and 11) OR (EXTRACT(HOUR FROM start_date) BETWEEN 15 and 21)) AND (EXTRACT(dayofweek FROM start_date) BETWEEN 2 and 6) GROUP BY subscriber_type' > commuterchart.csv
df_commuterchart = pd.read_csv('commuterchart.csv')
```

```python
df_commuterchart2= df_commuterchart.set_index('subscriber_type')
df_commuterchart2.plot.bar(color = 'green')
plt.xlabel('Subscriber Type')
plt.ylabel('Count of Commuter Trips Per Weekday')
plt.legend()
```

Total: ~150,000 commuter trips per weekday.
The vast majority of commuters in a given weekday are of type subscriber. Proposal 1 hopes to directly target them and increase their presence during the weekend.

For proposal part 2, we can see that this pass is particularly impactful given customer ride duration on weekends.

```python
! bq query --use_legacy_sql=false --format=csv 'SELECT floor((duration_sec/60)/5.00)*5 as bucket_floor, count(*) as counts, MAX(EXTRACT(hour FROM start_date) BETWEEN 2 AND 6) as weekday FROM `bigquery-public-data.san_francisco.bikeshare_trips` group by bucket_floor, (EXTRACT(hour FROM start_date) BETWEEN 2 AND 6) order by bucket_floor' > duration_trip.csv
df_duration = pd.read_csv('duration_trip.csv')
```

```python
df_duration['counts'] = np.log(np.array(df_duration['counts']))
ax = df_duration.groupby(['bucket_floor','weekday']).mean().unstack().plot.bar()
plt.xlabel('Trip Duration (in minutes)')
plt.ylabel('Log of Trip Count')
arr = np.array([])
for i in range(0,251,10):
    arr = np.append(arr, [i, ''])
L=plt.legend()
L.get_texts()[0].set_text('Weekday')
L.get_texts()[1].set_text('Weekend')
_ = ax.set_xticklabels(arr)
```

We can see a clear drop close to the 30 minute mark for ridership. The benefit of a weekend pass is that customers would not be limited to 30 minute rides as they currently are under the 24 hour or 3 day plan. Some customers already spend longer than 30 minutes riding for additional charges, but perhaps removing this charge on weekends would be helpful to increase ridership duration on weekends. This would incentivize longer rides (increased duration of trips) and also promote weekend travel.

To quantify the potential upshot from this proposal if both:

Number of Commuter trips = `150,000`

Total number of weekend trips (per day): `55,000`

If we were able to get the number of weekend trips on a given day to be equal to the number of just commuter trips on a given weekday, by applying these two policy proposals of a weekend pass and discount rates for subscribers who trip on weekends, we would see an increase in ridership of `200,000` 

100,000 difference per day * 2 weekend days = potentially `200,000` additional total trips.

In conclusion, these moderate proposals aim to increase ridership by focusing on improving weekend ridership, a currently massive gap via creative promotion of weekend riding via credits for future subscriptions and a weekend riding pass not capped at 30 minutes.
